## Q3. [Index] S&P 500 Market Corrections Analysis

Calculate the median duration (in days) of significant market corrections in the S&P 500 index.

For this task, define a correction as an event when a stock index goes down by more than 5% from the closest all-time high maximum.

- Download S&P 500 historical data (1950-present) using yfinance
- Identify all-time high points (where price exceeds all previous prices)
- For each pair of consecutive all-time highs, find the minimum price in between
- Calculate drawdown percentages: (high - low) / high × 100
- Filter for corrections with at least 5% drawdown
- Calculate the duration in days for each correction period
- Determine the 25th, 50th (median), and 75th percentiles for correction durations

```python
sp500_history = yf.download("^GSPC", start="1950-01-01")
sp500_history
```

```python
sp500 = sp500_history["Close"].reset_index()
sp500.columns = ["Date", "Close"]
sp500
```

```python
# Find out the all time high price as of each day
sp500["RollingMax"] = sp500["Close"].cummax()
sp500["IsCloseAllTimeHigh"] = sp500["Close"] > sp500["RollingMax"].shift(1)
sp500
```

```python
sp500_all_time_highs = sp500[sp500["IsCloseAllTimeHigh"]][["Date", "Close"]]
sp500_all_time_highs
```

```python
# Find minimum prices between all-time high dates
min_prices = []
min_duration = timedelta(days=2)

for i in range(len(sp500_all_time_highs) - 1):
    start_date = sp500_all_time_highs["Date"].iloc[i]
    end_date = sp500_all_time_highs["Date"].iloc[i + 1]

    if end_date - start_date < min_duration:
        continue

    # Filter data between two all-time high dates
    min_price = sp500[(sp500["Date"] > start_date) & (sp500["Date"] < end_date)]['Close'].min()

    min_prices.append({
        "Start_Date": start_date,
        "End_Date": end_date,
        "Min_Price": min_price
    })

min_prices = pd.DataFrame(min_prices)
min_prices
```

## Q4. [Stocks] Earnings Surprise Analysis for Amazon (AMZN)

Calculate the median 2-day percentage change in stock prices following positive earnings surprises days.

Steps:

1. Load earnings data from CSV (ha1_Amazon.csv) containing earnings dates, EPS estimates, and actual EPS. Make sure you are using the correct delimiter to read the data, such as in this command python pandas.read_csv("ha1_Amazon.csv", delimiter=';') 
2. Download complete historical price data using yfinance
3. Calculate 2-day percentage changes for all historical dates: for each sequence of 3 consecutive trading days (Day 1, Day 2, Day 3), compute the return as Close_Day3 / Close_Day1 - 1. (Assume Day 2 may correspond to the earnings announcement.)
4. Identify positive earnings surprises (where "actual EPS > estimated EPS" OR "Surprise (%)>0")
5. Calculate 2-day percentage changes following positive earnings surprises. Show your answer in % (closest number to the 2nd digit): return * 100.0
6. (Optional) Compare the median 2-day percentage change for positive surprises vs. all historical dates. Do you see the difference?

Context: Earnings announcements, especially when they exceed analyst expectations, can significantly impact stock prices in the short term.

```python
# amazon_earnings_data_url = "https://raw.githubusercontent.com/DataTalksClub/stock-markets-analytics-zoomcamp/refs/heads/main/cohorts/2025/ha1_Amazon.csv"
# amazon_earnings = pd.read_csv(amazon_earnings_data_url, delimiter=";")

amazon_earnings = pd.read_csv("homework_data/ha1_Amazon.csv", delimiter=";", na_values="-")
amazon_earnings
```

```python
amazon_price = yf.download("AMZN")
amazon_price
```

```python
amazon_price.columns = amazon_price.columns.droplevel(1)
```

```python
# 2-day percentage return = Day 3 close - Day 1 close
amazon_price["Return2D"] = amazon_price["Close"].diff(periods=2) / amazon_price["Close"]
amazon_price
```

```python
# Positive earnings surprises (where "actual EPS > estimated EPS" OR "Surprise (%)>0")
positive_surprises = (amazon_earnings["Reported EPS"] > amazon_earnings["EPS Estimate"]) | (amazon_earnings["Surprise (%)"] > 0)
positive_surprise_dates = amazon_earnings[positive_surprises]["Earnings Date"]
positive_surprise_dates
```

```python
# pd.to_datetime(positive_surprise_dates, format="%B %d, %Y at %I %p %Z")
positive_surprise_dates_parsed = pd.to_datetime(positive_surprise_dates, format="mixed").dt.date
```

```python
return_after_positive_surprise = amazon_price.loc[positive_surprise_dates_parsed + timedelta(days=1)]["Return2D"] * 100.0
return_after_positive_surprise
```

```python
return_after_positive_surprise.median()
```
